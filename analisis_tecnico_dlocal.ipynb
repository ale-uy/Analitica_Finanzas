# Análisis técnico de dLocal (uruguay)

Obtenemos de Yahoo Finance los datos históricos de OHLC (open, high, low & close), Adj close y volume de las acciones de dLocal (DLO), una fintech uruguaya considerada como la pionera uruguaya en el área, utilizando todos los datos disponibles en la plataforma

```python
# Importamos yahoo finanzas
#!pip install yfinance
import yfinance as yfin

# Datos de dLocal (DLO)
dlo = yfin.download('DLO', start='2020-01-01', end='2023-09-1') 
```

Utilizamos la librería de Python mplfinance para hacer un gráfico de Velas Japonesas para la acción de dLocal (DLO). 

```python
# Importamos mlpfinance
#!pip install mplfinance
import mplfinance as mpf

mpf.plot(dlo, type='candle', volume=True, figratio=(19,8), style='yahoo', title='dLocal')
```

Aparentemente estamos tomando un horizonte bastante amplio de tiempo que nos impide ver claramente las Velas Japonesas. Filtraremos los datos de entre el 2022-10-01 hasta el 2022-12-01 y analizaremos la reacción del mercado a finales del Covid 19.  

```python
mpf.plot(dlo["2022-10-01":"2022-12-01"], type='candle', volume=True, figratio=(19,8), style='yahoo', title='dLocal') 
```

Nos interesa analizar de manera detallada un día atipico que tuvo dLocal el 2022-11-15 hasta el 2022-11-17 con velas de 1 hora, no diarias.

```python
dlo_1 = yfin.download('DLO', start='2022-11-15', end='2022-11-18', interval ="1h") 
```

Grafíco las velas japonesas de la nueva serie de dLocal para esos días y agrego una media móvil de 5 y 20 horas mediante el parámetro mav.

```python
mpf.plot(dlo_1, type='candle', volume=True, figratio=(19,8), style='yahoo', title='dLocal', mav=(5, 20)) 
```
