# Anlizemos el valor de cierre de dLocal con un modelo Prophet

```python
# Importamos librerias
import pandas as pd
import matplotlib.pyplot as plt
from vx import Profeta
```

Cargamos los datos que fueron guardados como csv (dlocal_data.csv) desde el archivo dlocal_arima.ipynb

```python
dlocal = pd.read_csv('dlocal_data.csv')
dlocal['Date'] = dlocal['Date'].str.split().str[0]
```

```python
dlocal
```

## Aplicamos Modelo Prophet

```python
modelo_profeta = Profeta.entrenar_modelo(dlocal, target="Close", dates="Date", grid=True)
```

Obtenemos los ultimos datos para predecir y graficar

```python
dlocal_new = pd.read_csv('dlocal_new.csv')
dlocal_new['Date'] = dlocal_new['Date'].str.split().str[0]
```

Predecimos 5 valores a futuro

```python
# Hacer predicciones con el modelo
future = modelo_profeta.make_future_dataframe(periods=5)
forecast = modelo_profeta.predict(future)
```

Graficamos los resultados

```python
# Visualizar los datos originales y las predicciones
plt.figure(figsize=(12, 6))
plt.plot(dlocal_new.index, dlocal_new['Close'], label='Datos originales', color='blue')
plt.plot(dlocal_new.index, forecast['yhat'][-5:], label='Predicciones Prophet', color='red')
plt.legend(loc='best')
plt.title('Predicciones Prophet')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.show()
```
